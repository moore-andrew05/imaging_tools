```python
def plot_barcodes(barcodes, titles, save=None):

        fig, axes = plt.subplots(len(barcodes), 1)


        for i, bar in enumerate(barcodes):

            axes[i].imshow(bar, cmap="Greys_r", vmin = 0, vmax=1.0)
            axes[i].set_title(titles[i])
            axes[i].axis("off")

        fig.tight_layout(pad=0)
        
        
        if save is not None:
            fig.savefig(save, dpi=1200)
```

```python
plot_barcodes(bar_avgs, ["JuB66(RFP)", "JuB66(RFP) in CEMBIO", "Methanol Killed JuB66(RFP)"], "./REP1_IMAGES/Averages.svg")
```
